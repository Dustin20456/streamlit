```python
df2 = pd.read_csv ('./Downloads/life_expectancy_years.csv')
df2
```

```python
# show rows with null values
df2[df2.isna().any(axis=1)]
```

```python
# forward filling of missing values, excluding the countries column as the country name would be inserted if
# there is a missing value in the 1800 column
df2_1 = df2.iloc[:,1:].ffill(axis=1)
df2_1
```

```python
#backward filling of missing values
df2_1 = df2_1.bfill(axis ='columns')

df2_1
```

```python
df2_2 = df2['country']
```

```python
#merge it together with the countries again to one dataframe
df2_m = pd.merge(df2_2, df2_1, how='inner', left_index=True, right_index=True)

df2_m
```

```python
#check if all null values are gone now
df2_m[df2_m.isna().any(axis=1)]
```

```python
#transfer to tidy data format
df2_m = pd.melt(df2_m,id_vars=['country'])
df2_m
```

```python
# sort rows and rename colums
df2_m = df2_m.sort_values(by=['country', 'variable'])
df2_m = df2_m.rename(columns={'variable': 'year', 'value':'life expectancy'})
df2_m
```

```python
#export dataframe to csv file
df2_m.to_csv ('./Downloads/life_expectancy.csv', index = False, header=True)
```
